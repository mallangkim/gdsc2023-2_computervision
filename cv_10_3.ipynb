```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import requests
from io import BytesIO
```

```python
url = 'https://search.pstatic.net/common?type=f&size=518x522&quality=95&direct=true&src=http%3A%2F%2Fshop1.phinf.naver.net%2F20200217_150%2F1581909134674UIiu7_JPEG%2F18044751731076648_279309417.jpg'

response = requests.get(url)
pic = Image.open(BytesIO(response.content))
pic
```

- opencv에서 색을 읽는 방법
 - RBG (X)

```python
image = cv2.cvtColor(pic_arr, cv2.COLOR_BGR2RGB)
cv2_imshow(image)
```

```python
copy2 = pic_copy
img = cv2.cvtColor(copy2, cv2.COLOR_BGR2RGB)
# image = cv2.cvtColor(pic_arr, cv2.COLOR_BGR2RGB)
cv2_imshow(img)
```
